```python
import pandas as pd
```

```python
unames = ['user_id', 'Gender', 'Age', 'occupation', 'zip']
users = pd.read_csv('users.dat', sep='::', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv('ratings.dat', sep='::', header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_csv('movies.dat', sep='::', header=None, names=mnames)
```

```python
users[:5]
```

```python
movies[:5]
```

```python
ratings[:5]
```

```python
#Pandas Merge function
#Ratings + users + Movies
data = pd.merge(pd.merge(users, ratings),movies)
data.head()
```

```python
data.iloc[0]
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='Gender', aggfunc='mean')
```

```python
print("Calificacion media de peliculas divida por genero")
mean_ratings[:5]
```

```python
#Agrupacion por titulo de pelicula y cantidad de veces registrada.
raiting_by_title = data.groupby('title').size()
raiting_by_title[:10]
```

```python
#Filtra datos 
#solo quiero los films que hayan sido calificados mas de 200 veces.
active_title = raiting_by_title.index[raiting_by_title>200]
active_title[:5]
```

```python
#puedo usar mi filtro previo
#calificacion media de las peliculas calificadas mas de 200 veces
#dividido por genero
mean_ratings = mean_ratings.loc[active_title]
mean_ratings[:10]
```

```python
#Ver las peliculas que fueron mejor calificadas por las mujeres
#ascending=False => Mayor a menor valor
top_famale_raiting = mean_ratings.sort_values(by='F', ascending=False)
top_famale_raiting[:10]
```

```python
#Hallado las peliculas mas divisadas entre hombres y mujeres.
#Diferencia de clasificación entra ambos sexos.(diff)

mean_ratings['diff']= mean_ratings['M']-mean_ratings['F']

mean_ratings[:10].sort_values(by='diff', ascending=False)
```

```python

sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:10]
```

```python
#Reversion en el orden de las filas, top 15 films prerefidos de los hombres
sorted_by_diff[::-1][:15].sort_values(by='M', ascending=False)
```

```python
#Film peor calificado, independientemente del genero
#Desciacion estandar de ratings agrupado por titulo

rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title.head(10)
```

```python
#Filtro films, solo las que tienen mas de 200 calificaciones.
rating_std_by_title = rating_std_by_title.loc[active_title]
rating_std_by_title.head()
```

```python
rating_std_by_title.sort_values(ascending=False)[:10]
```
