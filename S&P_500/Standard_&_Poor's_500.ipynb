```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
from datetime import datetime
#%matplotlib notebook
```

```python
data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
data.head()
```

```python
fig, ax = plt.subplots()
spx.plot(ax=ax, grid=True, style='k-')


crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date)+5),
               xytext=(date, spx.asof(date)+150),
               arrowprops=dict(facecolor='red'))
              

#zoom en 2007-2009
ax.set_xlim(['30/7/2007', '30/3/2009'])
ax.set_ylim([600, 1800])

ax.set_title('Fechas importantes entre 2007 - 2008 en las crisis financieras',
            fontdict={'fontsize': 10,
                 'fontweight' :'extra bold',
                 'verticalalignment':'baseline'})
```
